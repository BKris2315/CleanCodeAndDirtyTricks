# II. Meaningful Names

## üß† Why Names Matter

- Naming is **one of the hardest problems** in programming.
- Poor naming leads to **confusion, bugs, and harder maintenance**.
- Your code is read more often than it‚Äôs written ‚Äî name things like you care!

If your naming is bad, you need comments. **A well-written code does not need any additional comment** apart from a docstring. More on comments in Ch. IV.


## üéØ Use Intention-Revealing Names
<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">

```c

```
</div>

<div style="width: 48%;">

**Good:**
```C

```
</div>
</div>

### > Names should clearly describe what the variable does or represents.

<div style="display: flex; justify-content: space-between;">

### > Misleading names are worse than vague ones ‚Äî don‚Äôt lie to the reader.

## üßæ Mainstream Naming Conventions

| Case Style        | Used For                           | Example                    | Common In                   |
|-------------------|-------------------------------------|----------------------------|-----------------------------|
| **camelCase**     | Variables, functions                | `totalCount`, `saveUser()` | JavaScript, Java, C#, Swift |
| **PascalCase**    | Classes, types, constructors        | `CustomerAccount`, `AppView` | C#, Java, TypeScript       |
| **snake_case**    | Variables, functions (esp. in Python) | `total_count`, `save_user()` | Python, Ruby, C, C++        |
| **SCREAMING_SNAKE_CASE** | Constants                     | `MAX_CONNECTIONS`, `DEFAULT_TIMEOUT` | C, Python, Go, many more   |
| **kebab-case**    | Filenames, URLs, CLI options        | `clean-code.md`, `--user-name` | HTML/CSS files, URLs, CLI  |
| **dot.case**      | Config files, package names         | `clean.code.config`        | JSON, npm, file systems     |


### ‚úÖ General Naming Best Practices

- **Be consistent**: Stick to one convention per project.
- **Reflect meaning**: Names should describe purpose, not implementation.
- **Avoid abbreviations** unless widely accepted (e.g. `id`, `url`).
- **Pluralize collections**: `users` for a list, `user` for a single object.
- **Prefix booleans** with verbs: `isVisible`, `hasAccess`, `canEdit`.

## Summary

<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">

- Reveal intent
- Be specific
- Use consistent terms
- Favor domain terminology

**üö´ Avoid:**
- Vague, short names
- Misleading labels
- Arbitrary distinctions
- Jargon or abbreviations


</div>

<div style="width: 48%;">

### > ‚ÄúNaming is not just about identification ‚Äî it‚Äôs about communication.‚Äù

Well-named code explains itself and reduces the need for comments. It‚Äôs one of the most powerful tools in writing clean, professional software.

# III. Functions
_A guide to writing clean, small, and meaningful functions_

<img src="imgs/programmerhumor-io-python-memes-backend-memes-6171778fb27c5f2.jpg" alt="Drawing" style="height: 500px;"/>

## üìè Keep It Small

> ‚ÄúFunctions should be short. Really short."

**Why?**  
- Easier to read and understand  
- Easier to reuse and test  
- Tells a story, one step at a time

### Example 

<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">


### ‚õîÔ∏è **Bad**
```python
def process_user(user_data):
    # many unrelated operations here
    ...
```
</div>

<div style="width: 48%;">

## ‚úÖ **Good**
```python
def validate_user(user_data): ...
def store_user(user_data): ...
def notify_user(user_data): ...
```
</div>
</div>

---

<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">

```python
def elapse(year):
    days = 365
    if year % 4 == 0 or (year % 100 == 0 and year % 400 == 0):
        days += 1
    for day in range(1, days + 1):
        print(f"Day {day} of {year}")
```
</div>

<div style="width: 48%;">

```python
def is_leap(year): ...
def elapse(year):
    days = 366 if is_leap(year) else 365
    ...
```
</div>
</div>


## üéØ Do One Thing

A function should do **only one thing** ‚Äî and do it well. Also called the **DRY Principle** (Don't Repeat Yourself).

<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">

### ‚õîÔ∏è Bad (Python):
```python
def create_user(name, email, password):
    if not name:
        raise ValueError("Invalid name")
    store_user_in_db(name, email, hash_password(password))
```
</div>

<div style="width: 48%;">

### ‚úÖ Good (Python):
```python
def validate_name(name): ...
def hash_password(password): ...
def store_user(name, email, hashed_pw): ...
```
</div>
</div>

## üß† Use Descriptive Names

**Good names tell the reader what the function does.**
<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">

### ‚õîÔ∏è Bad (C):
```c
int doIt(int a, int b) {
    return a + b;
}
```
</div>

<div style="width: 48%;">

### ‚úÖ Good (C):

```c
int addTwoNumbers(int firstNumber, int secondNumber) {
    return firstNumber + secondNumber;
}
```
</div>
</div>

## üßÆ Minimize Function Arguments

<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">

- Ideal: **0 (niladic)**  
- Good: **1 (monadic)**  
- Acceptable: **2 (dyadic)**  
- Avoid: **3+ (triadic/polyadic)** unless necessary
</div>

<div style="width: 48%;">

### üßµ Bundle related args:
```python
# Instead of:
def draw_box(x, y, width, height): ...

# Prefer:
def draw_box(rect):  # where rect is a namedtuple or object
```
</div>
</div>


## ‚ùì Command‚ÄìQuery Separation

> Functions should either **do something** or **answer something**, not both.

<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">

### ‚õîÔ∏è Bad:
```python
def set_username(username):  # returns success?
    ...
```
</div>

<div style="width: 48%;">

### ‚úÖ Good:
```python
def username_exists(username): ...
def update_username(username): ...
```
</div>
</div>


## üö® Prefer Exceptions Over Error Codes

<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">

### ‚õîÔ∏è Bad (C-style):

```c
int openFile(char* path) {
    if (path == NULL) return -1;
    // ...
    return 0;
}
```
</div>

<div style="width: 48%;">

### ‚úÖ Good (Python):
```python
def open_file(path):
    if not path:
        raise ValueError("Invalid path")
```
</div>
</div>

## üîÅ Don't Repeat Yourself (DRY), again

Avoid copy-pasting logic ‚Äî abstract instead.
<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">

### ‚õîÔ∏è Bad (Python):
```python
def sum_two(a, b): return a + b
def sum_three(a, b, c): return a + b + c
```
</div>

<div style="width: 48%;">

### ‚úÖ Good:
```python
def sum_all(*args):
    return sum(args)
```
</div>
</div>


## ‚úÖ Recap: Function Best Practices

| Principle | Description |
|----------|-------------|
| **Small** | Short and focused |
| **Do One Thing** | No multitasking |
| **Descriptive Names** | Self-documenting code |
| **Few Arguments** | Prefer 0‚Äì2 args |
| **Command ‚â† Query** | Clear responsibility |
| **Use Exceptions** | Don‚Äôt return error codes |
| **DRY** | Reuse logic, avoid duplication |


### > _‚ÄúClean functions are like clean tools: precise, sharp, and made for a single task.‚Äù_

Let your functions tell a story ‚Äî clearly, one well-named step at a time.


# IV. Comments Are Bad
_When (not) to comment your code_

<img src="imgs/stop-sign-writing-comments-stop-this-is-stop-sign.jpeg" alt="Drawing" style="height: 500px;"/>


## üßº The Philosophy

> _‚ÄúDon‚Äôt comment bad code ‚Äî rewrite it.‚Äù_  
‚Äî Robert C. Martin

- Comments are a **necessary evil**, not a best practice.
- Code should be **self-explanatory** ‚Äî comments are a fallback.
- Most comments decay into **misleading noise** over time.

## ‚ö†Ô∏è Why Comments Can Be Bad

- **Get outdated** as code changes
- Can **lie** or mislead
- Are often used to explain **unclear code**
- Create **maintenance burden**


## üß† Code Should Speak For Itself

<div style="display: flex; justify-content: space-between;">

<div style="width: 48%;">

Instead of this:
```python
# increase balance by 10%
balance = balance + balance * 0.1
```
</div>

<div style="width: 48%;">

Do this
```python
def apply_interest(balance, rate=0.1):
    return balance * (1 + rate)
```
</div>
</div>



## ‚úÖ When Comments Are Useful

### üìú Legal Comments  
Licensing and copyright info.

### üßæ Informative Comments  
Explaining non-obvious things (e.g., regexes, hardware constraints).

### üß† Intent Explanation  
Why something is done a certain way, not *what* is being done.

```python
# Priority queue used for Dijkstra's algorithm
queue = PriorityQueue()
```


### üîç Clarification  
For unavoidable obscurities:
```python
# API returns 0 for success, 1 for failure ü§∑
if api_call() == 0: ...
```

### ‚ö†Ô∏è Warning of Consequences  
```python
# ‚ö†Ô∏è This test takes ~5 min to run!
def test_large_dataset(): ...
```

### üõ† TODOs  
Track incomplete tasks (but remove when resolved).

```python
# TODO: Add support for edge cases
```

### üì£ Amplification  
Highlighting subtle but crucial detail:
```python
# Trim to avoid trailing whitespace bugs
cleaned_name = name.strip()
```

### üìö Javadocs / Public API Docs  
Needed for external consumers, but must be maintained.



## üö´ Bad Comment Patterns

| Type               | Problem                                |
|--------------------|-----------------------------------------|
| **Mumbling**       | Vague, unclear wording                  |
| **Redundant**      | Repeats what the code already shows     |
| **Misleading**     | No longer accurate                      |
| **Mandated**       | Required but meaningless docs           |
| **Journal Comments** | Version history ‚Äî use Git instead    |
| **Noise**          | Adds clutter, not clarity               |


## ‚ùå Examples of Bad Comments

```python
# increment i by 1
i += 1  # ‚Üê no need for that comment
```

```c
// Default constructor
MyClass::MyClass() {}
```

```java
// Changed on Jan 2023
// Added null check
```
> Use version control (Git!) instead.

## üßº Comment TL;DR

| üí° Use Comments For         | üö´ Avoid Comments That      |
|-----------------------------|-----------------------------|
| Legal / license info        | State the obvious           |
| Intent or rationale         | Get out-of-sync easily      |
| Real warnings / TODOs       | Replace poor naming         |
| Public API documentation    | Could be refactored instead |

### > _‚ÄúThe best comment is the one you didn‚Äôt need to write because your code was clear enough.‚Äù_

# üìÑ Writing Good Docstrings (Google Style)

> _‚ÄúComments explain **why**, docstrings explain **how to use**.‚Äù_

## üß† Why Use Docstrings?

- Clarify usage of your functions/classes/modules
- Help others (and future you) understand API contracts
- Power tools like `help()`, IDEs, and doc generators

## ‚úÖ Google Style Format

```python
def get_user(id: int, include_inactive: bool = False) -> dict:
    """
    Fetch a user by ID.

    Args:
        id (int): Unique user ID.
        include_inactive (bool): Include inactive users. Default is False.

    Returns:
        dict: User data dictionary.

    Raises:
        ValueError: If ID is not found or invalid.
    """
```

### ‚ú® Tips for Great Docstrings

- First line = **short summary** (ends with period)
- Use **imperative tone** (‚ÄúReturn...‚Äù not ‚ÄúReturns‚Ä¶‚Äù)
- Include `Args`, `Returns`, `Raises`, and optional `Examples`
- Document non-obvious behavior, defaults, edge cases

## üß© Why Use Virtual Environments?

- Avoid version conflicts between projects
- Reproducible environments for collaboration or deployment
